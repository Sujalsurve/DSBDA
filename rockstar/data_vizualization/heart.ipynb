```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 5: Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 6: Target variable distribution
sns.countplot(data=df, x='num', palette='Set2')
plt.title('Distribution of Heart Disease (Target)')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Step 7: Age distribution by heart disease
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='num', multiple='stack', bins=20, palette='Set1')
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Step 8: Chest pain type vs. target
sns.countplot(data=df, x='cp', hue='num', palette='Set3')
plt.title('Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# Step 9: Cholesterol vs Age scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='num', palette='cool')
plt.title('Cholesterol vs Age')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# Step 10: Boxplot of resting blood pressure by target
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='num', y='trestbps', palette='pastel')
plt.title('Resting Blood Pressure by Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```
